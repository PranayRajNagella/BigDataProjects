```
#String Palindrome
mystring=str(input())
copyofString=mystring
print(mystring[::-1])
if(mystring==copyofString):
    print(1)
else:
    print(0)
```

```
#ReverseWords
mysentence=str(input())
words=mysentence.split(" ")
finalsentence=""

for word in words:
    finalsentence=finalsentence+word[::-1]+' '

print(finalsentence)
```
